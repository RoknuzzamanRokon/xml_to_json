# Creating json format

### Import libarey for create json file 

```python
import requests
import xmltodict
import json
```

### Define the URL for the XML data

```python
url = "https://affiliatefeed.agoda.com/datafeeds/feed/getfeed?apikey=3ce93b4c-3bb4-4f0f-91c9-6e861c2bf7fb&mhotel_id=15346000&feed_id=19"

# Fetch the XML data from the URL
response = requests.get(url)
```

```python
if response.status_code == 200:
    # Parse XML to OrderedDict
    xml_data = response.content
    data_dict = xmltodict.parse(xml_data)
    
    # Convert OrderedDict to JSON
    json_data = json.dumps(data_dict, indent=4)
    
    # Print JSON data
    print(json_data)
else:
    print("Failed to retrieve XML data")
```
